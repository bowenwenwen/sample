```python
adp_clock_data['stack'] = adp_clock_data['server_name'].str.split('-').str[0]

adp_clock_data['clock_enabled'] = np.where(
    adp_clock_data['clock_enabled'] == "f",
    0,
    1
)

adp_clock_data['biometric_consent_required'] = np.where(
    adp_clock_data['biometric_consent_required'] == False,
    0,
    1
)

adp_clock_data["report_run_date"] = report_run_date
```

```python
adp_clock_data["report_run_date"] = pd.to_datetime(adp_clock_data["report_run_date"]).dt.date
```
